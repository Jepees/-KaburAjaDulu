```python
import os
import pandas as pd
import numpy as np
```

```python
path = "./tweets-data"
datas = os.listdir(path)
datas1 = [i for i in datas if "Kabur" in i]
# datas2 = [i for i in datas if "Kabur" not in i]
datas1
```

```python
dfs = {f"df{i}": pd.read_csv(f"{path}/{datas1[i]}") for i in range(len(datas1))}
df1 = pd.concat(list(dfs.values()), ignore_index=True)
df1['Media'] = df1.shape[0]*["X"]
df1
```

```python
df1[df1['user_id_str'] == 3711470718]['tweet_url'].iloc[0]
```

```python
print(f"jumlah data duplikat sebelum dihapus: {df1.duplicated().sum()}")
df1 = df1.drop_duplicates()
print(f"jumlah data duplikat setelah dihapus: {df1.duplicated().sum()}")
```

```python
df1 = df1.rename(columns={'favorite_count': 'like_count'})
df1 = df1[[ 'created_at', 'user_id_str', 'full_text', 'like_count', 'quote_count', 'reply_count', 'retweet_count', 'tweet_url', 'Media' 
        # 'id_str','username',
        ]]
df1.columns = ['created_at', 'username/id_user', 'full_text', 'like_count', 'quote_count', 'reply_count', 'retweet_count', 'tweet_url', 'Media']
df1['username/id_user'] = df1['username/id_user'].astype(int)
df1
```

```python
df2 = pd.read_csv(f"{path}/AMS scraping yt.csv")
df2['quote_count'] = df2.shape[0]*[0]
df2['reply_count'] = df2.shape[0]*[0]
df2['retweet_count'] = df2.shape[0]*[0]
df2['tweet_url'] = df2.shape[0]*[np.nan]
df2['Media'] = df2.shape[0]*["YouTube"]
df2
```

```python
df2.columns = ['created_at', 'username/id_user', 'full_text', 'like_count', 'quote_count', 'reply_count', 'retweet_count', 'tweet_url', 'Media']
df2
```

```python
dff = pd.concat([df1, df2], ignore_index=True)
dff
```

```python
dff.to_csv(f"{path}/data_UTS_fix.csv", index=False)
```
